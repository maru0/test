## GAコード


```
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import pdb
import collections
```

```
class GA():
  def __init__(self):
    self.M = 20 # 個体数
    self.D = 5 # 種類数
    self.Pm = 0.05
    self.Tmax = 100
    self.Weight = np.zeros(self.D)
    self.Value = np.zeros(self.D)
    #self.Wmax = float("inf")
    self.Wmax = 15

    #self.X = np.random.rand(self.M,self.D)
    self.X = np.random.randint(0,2,(self.M,self.D))
    self.Xnext = np.zeros((self.M,self.D))
    self.F = np.zeros(self.M)
    self.Fbest = 0
    self.Xbest = np.zeros(self.D)
    self.Fbest_list = []
    self.Xbest_list = []
    self.p1 = 0
    self.p2 = 0
    self.d1 = 0
    self.d2 = 0


  def simulation(self):
    t = 0
    xtmp = 0
    ftmp = 0
    while(t < self.Tmax):
      t = t+1
      self._target_update()
      for best_type in range(self.M):
        #if np.any(xtmp > self.X[best_type])):
        #  xtmp = self.X[best_type]
        if np.any(ftmp < self.F[best_type]):
          self.Xbest = self.X[best_type]
          ftmp = self.F[best_type]
      #self.Xbest = xtmp
      #self.Xbest_list.append(self.Xbest)
      self.Fbest = ftmp
      self.Fbest_list.append(str(self.Fbest))
      p2_F = self.F
      for val_type in range(self.M):
        r = np.random.rand()
        for i in range(self.M):
          # ルーレット選択
          prob = self.F[i] / sum(self.F)
          if r < prob:
            self.p1 = i
            np.delete(p2_F,i)
            break
          # ここ問題かも
          r -= prob

        r = np.random.rand()
        for i in range(i,self.M):
          if i != self.p1:
            prob = self.F[i] / sum(p2_F)
            if r < prob:
              self.p2 = i
              break
            r -= prob

        self._two_crossing(val_type)
      self.X = self.Xnext
    self._ans_print()
    return self.Fbest_list,self.Xbest_list

  # 評価関数
  def _target_update(self):
    for i in range(self.M):
      self.F[i] = 0
      tmp = 0
      for dimension in range(self.D):
        self.F[i] += self.Value[dimension] * self.X[i][dimension]
        tmp += self.Weight[dimension] * self.X[i][dimension]
        #self.F[dimension] = self.Value[dimension]
      if tmp > self.Wmax:
        self.F[i] = 1.0


  def _two_crossing(self,val_type):
    dimension_list = np.array([value for value in range(self.D)])
    self.d1,self.d2 = np.random.choice(dimension_list,2,replace=False)
    if self.d1 > self.d2:
      self.d1,self.d2 = self.d2,self.d1
    for dimension in range(self.D):
      if dimension <= self.d1 or dimension > self.d2:
        self.Xnext[val_type][dimension] = self.X[self.p1][dimension]
      else:
        self.Xnext[val_type][dimension] = self.X[self.p2][dimension]
    for dimension in range(self.D):
      if np.random.rand() < self.Pm:
        self.Xnext[val_type][dimension] = 1 - self.Xnext[val_type][dimension]
        
  def _ans_print(self):
    print("解の目的関数値Fg ={}".format(max(self.Fbest_list)))
    print("最適解 Xbest:")
    #for dimension in range(self.D):
    #print(self.Xbest_list[dimension])
    print(self.Xbest)
    #c = collections.Counter(self.Fbest_list)
    #print(c.most_common()[0])
```

```
hoge = GA()
hoge.Wmax=15
hoge.M = 20
hoge.D = 5
hoge.Value = [50,40,10,70,55]
hoge.Weight = [7,5,1,9,6]
hoge.simulation()
```
